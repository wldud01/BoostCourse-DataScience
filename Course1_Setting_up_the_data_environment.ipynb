```python
import pandas as pd
# pd.dataFrame?
# pd.DataFrame(shift + tab + tab)
df = pd.DataFrame(
{"a" : [4, 5, 6],
"b" : [7, 8, 9],
"c" : [10, 11, 12]},
index = [1, 2, 3])
df
```

```python
# Series , vector
df['a']
```

```python
# dataframe
df[['a']]
```

```python
# boolean indexing
df[df['a'] >4]
```

```python
# columns의 값을 list 형태로 전달해서 값을 여러개 사용할 수 있음
df[['a','c']]
```

```python
# summarize data
print("uniuque value")
print(df['a'].value_counts())
print("-----------------------------------------")
print("distict value in a column")
print(df['a'].nunique())
print("-----------------------------------------")
print("basic statistics for each columns")
print(df.describe())
```

```python
df['a'].sort_values()
```

```python
# columns 값을 기준으로 정렬
df.sort_values("a",ascending=False)
```

```python
# drop을 할때 axis로 행과 열 중 어떤 것을 기준으로 정렬할지 잘 해주어야 한다
df.drop(['c'])
```

```python
print(df.groupby(['a'])['b'].mean())
print(df.groupby(['a'])['b'].agg(['mean','sum','count']))
```

```python
df.groupby(['a'])['b'].describe()
```

```python
pd.pivot_table(df,index="a",values='b')
```

```python
# 경로 안에 포함되는 경로 찾기
from glob import glob
```

```python
import pandas as pd

#file 경로
file_name = glob('drive/MyDrive/data/*')
file_name
```

```python
df = pd.read_csv(file_name[0],encoding='cp949')
```

```python
df.head()
```

```python
df.info()
```

```python
df['성별(GEDNER)'][df['성별(GEDNER)'] == 'F'] = 1
df.loc[df['성별(GEDNER)']=='M','성별(GEDNER)'] = 0
```

```python
df['성별(GEDNER)'] = df['성별(GEDNER)'].astype(int)
```

```python
df.describe()
```

```python
sex_count = df['성별(GEDNER)'].value_counts()
sex_count.plot.bar(color=['orange','green'])
```

```python
df.groupby(df['성별(GEDNER)'])['카드이용건수계(USECT_CORR)'].plot(legend=['F','M'])
```

```python
df.groupby('성별(GEDNER)')["카드이용금액계(AMT_CORR)"].sum().plot.bar(color=['orange','green'])
```

```python
df.groupby('성별(GEDNER)')["카드이용건수계(USECT_CORR)"].sum().plot.bar(color=['orange','green'])
```
